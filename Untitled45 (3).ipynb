```python
# ============================================================
# SNIPPET 1: SETUP + FEATURE ENGINEERING
# Run this once before running any model snippet
# ============================================================

import zipfile, os
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler


# --- Load ---
zip_path = "/content/archive (7).zip"
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall("/content/creditcard_data")
df = pd.read_csv("/content/creditcard_data/creditcard.csv")
print("Shape:", df.shape)
print(df["Class"].value_counts())

# --- Feature Engineering ---
# Why log transform? Amount is highly right-skewed; log compresses large values
df["Amount_log"] = np.log1p(df["Amount"])

# Fraud patterns often spike at certain hours of the day
df["Hour"] = (df["Time"] // 3600) % 24

# Z-score captures how unusual the transaction amount is
df["Amount_zscore"] = (df["Amount"] - df["Amount"].mean()) / df["Amount"].std()

# Interaction features: these PCA components are known fraud indicators;
# multiplying them lets the model capture joint behavior
df["V1_V2"]   = df["V1"]  * df["V2"]
df["V3_V4"]   = df["V3"]  * df["V4"]
df["V10_V14"] = df["V10"] * df["V14"]

# Raw Time is not useful after extracting Hour
df = df.drop(["Time"], axis=1)

# --- Split Features & Target ---
X = df.drop("Class", axis=1)
y = df["Class"]

# Scale only the non-PCA numeric columns (PCA columns are already scaled)
scaler = StandardScaler()
cols_to_scale = ["Amount", "Amount_log", "Hour", "Amount_zscore"]
X[cols_to_scale] = scaler.fit_transform(X[cols_to_scale])

# Stratify ensures both train and test have the same fraud ratio
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

print(f"\nTrain size: {X_train.shape}, Test size: {X_test.shape}")
print("Data ready! Now run either Model Snippet A or B.")

# We'll store results here so the comparison snippet can access them
model_results = {}  # dict to collect {model_name: {roc_auc, pr_auc, report}}
```

```python
# ============================================================
# SNIPPET 2A: RANDOM FOREST — Full Version (Optimized for Speed)
# Includes: Feature Importance + Cross Validation + Final Evaluation
# Run Snippet 1 before this
# ============================================================

from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import StratifiedKFold, cross_val_score
from sklearn.metrics import classification_report, roc_auc_score, average_precision_score
import matplotlib.pyplot as plt
import numpy as np
import time

print("=" * 55)
print("        RANDOM FOREST — FULL EVALUATION PIPELINE")
print("=" * 55)

# -----------------------------------------------------------
# MODEL DEFINITION
# Key speed optimizations explained below:
# - n_estimators=50   : 50 trees instead of 100/200. Beyond 50,
#                       accuracy gains are tiny but time doubles.
# - max_depth=12      : Without this, trees grow 25-30 levels deep
#                       on 280k rows — mostly memorizing noise.
#                       Capping at 12 is faster AND less overfit.
# - max_samples=0.6   : Each tree trains on only 60% of data.
#                       Speeds up each tree and adds diversity.
# - n_jobs=-1         : Uses ALL available CPU cores in parallel.
#                       On Colab (2 cores), this roughly halves time.
# -----------------------------------------------------------
rf = RandomForestClassifier(
    n_estimators=50,
    max_depth=12,
    max_samples=0.6,
    max_features="sqrt",
    min_samples_leaf=5,    # A node must have 5+ samples to split further
                           # prevents over-detailed splits on tiny groups
    class_weight="balanced",
    n_jobs=-1,
    random_state=42
)

# -----------------------------------------------------------
# STEP 1: CROSS VALIDATION
# We use 3 folds instead of 5 to cut CV time nearly in half.
# 3-fold on 227k training rows is still extremely reliable —
# each fold trains on ~151k samples which is more than enough.
# The mean ± std across 3 folds tells us if the model is stable.
# -----------------------------------------------------------
print("\n[1/3] Running Cross-Validation (3-fold)...")
skf = StratifiedKFold(n_splits=3, shuffle=True, random_state=42)

start = time.time()
cv_scores = cross_val_score(
    rf, X_train, y_train,
    cv=skf,
    scoring="roc_auc",
    n_jobs=-1   # Each fold also runs in parallel
)
cv_time = time.time() - start

print(f"      CV ROC-AUC : {cv_scores.mean():.4f} ± {cv_scores.std():.4f}")
print(f"      Fold Scores: {[round(s, 4) for s in cv_scores]}")
print(f"      Time Taken  : {cv_time:.1f}s")

# A tight std (e.g. 0.002) = model is stable across different data slices
# A wide std (e.g. 0.04)   = model is sensitive to which data it sees — red flag
if cv_scores.std() < 0.005:
    print("      Stability   : EXCELLENT (very consistent across folds)")
elif cv_scores.std() < 0.015:
    print("      Stability   : GOOD")
else:
    print("      Stability   : VARIABLE — consider tuning or more data")

# -----------------------------------------------------------
# STEP 2: FINAL TRAINING ON FULL TRAINING SET
# After CV confirms the model is good, we train one final time
# on ALL of X_train (not just 2/3 of it like in CV folds)
# so the model has seen as much data as possible before testing.
# -----------------------------------------------------------
print("\n[2/3] Training Final Model on Full Training Set...")
start = time.time()
rf.fit(X_train, y_train)
print(f"      Training Time: {time.time()-start:.1f}s")

# -----------------------------------------------------------
# STEP 3: EVALUATION ON HELD-OUT TEST SET
# This test set was never touched during training or CV —
# so these scores represent true unseen performance.
# -----------------------------------------------------------
print("\n[3/3] Evaluating on Test Set...")
y_pred_rf = rf.predict(X_test)
y_prob_rf = rf.predict_proba(X_test)[:, 1]

roc_rf = roc_auc_score(y_test, y_prob_rf)
pr_rf  = average_precision_score(y_test, y_prob_rf)

print("\n========== RANDOM FOREST — FINAL RESULTS ==========")
print(classification_report(y_test, y_pred_rf, digits=4))
print(f"  ROC-AUC : {roc_rf:.4f}   (how well model ranks fraud vs non-fraud)")
print(f"  PR-AUC  : {pr_rf:.4f}   ← trust this more for imbalanced fraud data")

# -----------------------------------------------------------
# BONUS: TOP 10 FEATURE IMPORTANCES
# Random Forest tells us which features contributed most to
# its decisions. This is useful for understanding the model
# and also for catching if something suspicious is dominating.
# -----------------------------------------------------------
print("\n--- Top 10 Most Important Features ---")
feat_importances = pd.Series(rf.feature_importances_, index=X_train.columns)
top10 = feat_importances.sort_values(ascending=False).head(10)

fig, ax = plt.subplots(figsize=(9, 5))
top10.sort_values().plot(
    kind='barh', ax=ax,
    color='steelblue', edgecolor='white'
)
ax.set_title("Random Forest — Top 10 Feature Importances", fontsize=13, fontweight='bold')
ax.set_xlabel("Importance Score")
ax.grid(axis='x', alpha=0.3)
plt.tight_layout()
plt.show()

for feat, score in top10.items():
    print(f"  {feat:<20} {score:.4f}")

# -----------------------------------------------------------
# SAVE RESULTS for the Comparison Snippet later
# -----------------------------------------------------------
model_results["Random Forest"] = {
    "cv_roc_auc"  : cv_scores.mean(),
    "cv_std"      : cv_scores.std(),
    "roc_auc"     : roc_rf,
    "pr_auc"      : pr_rf,
    "y_prob"      : y_prob_rf,
    "y_pred"      : y_pred_rf
}

print("\n Results saved to model_results['Random Forest']")
print(" Now run Snippet 2B to train XGBoost and then compare!")
```

```python
# ============================================================
# SNIPPET 2B: XGBOOST — Train, Evaluate, Store Results
# Run Snippet 1 before this (can run independently of 2A)
# ============================================================

from xgboost import XGBClassifier
from sklearn.model_selection import StratifiedKFold, cross_val_score
from sklearn.metrics import classification_report, roc_auc_score, average_precision_score
import time

# scale_pos_weight is XGBoost's way of handling class imbalance:
# it tells the model "each fraud sample counts this many times more"
neg  = (y_train == 0).sum()
pos  = (y_train == 1).sum()
spw  = neg / pos
print(f"scale_pos_weight = {spw:.1f}  ({neg} non-fraud vs {pos} fraud samples)")

xgb = XGBClassifier(
    n_estimators=200,          # Fewer trees = faster; increase for better performance
    max_depth=5,               # Shallower trees reduce overfitting
    learning_rate=0.1,
    scale_pos_weight=spw,      # Handles imbalance
    eval_metric='logloss',
    tree_method='hist',        # Faster histogram-based method
    n_jobs=-1,
    random_state=42
)

# --- Cross Validation ---
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

start = time.time()
cv_scores = cross_val_score(xgb, X_train, y_train, cv=skf, scoring="roc_auc", n_jobs=-1)
print(f"XGB Cross-Val ROC-AUC: {cv_scores.mean():.4f} ± {cv_scores.std():.4f}")
print(f"CV Time: {time.time()-start:.1f}s")

# --- Final Fit ---
xgb.fit(X_train, y_train)
y_pred_xgb = xgb.predict(X_test)
y_prob_xgb = xgb.predict_proba(X_test)[:, 1]

roc_xgb = roc_auc_score(y_test, y_prob_xgb)
pr_xgb  = average_precision_score(y_test, y_prob_xgb)

print("\n========== XGBOOST RESULTS ==========")
print(classification_report(y_test, y_pred_xgb, digits=4))
print(f"ROC-AUC : {roc_xgb:.4f}")
print(f"PR-AUC  : {pr_xgb:.4f}   ← more meaningful for imbalanced data")

# Store for comparison
model_results["XGBoost"] = {
    "cv_roc_auc": cv_scores.mean(),
    "roc_auc": roc_xgb,
    "pr_auc": pr_xgb,
    "y_prob": y_prob_xgb,
    "y_pred": y_pred_xgb
}
print("\nResults saved to model_results['XGBoost']")
```

```python
import numpy as np
from sklearn.metrics import f1_score

# Try every threshold from 0.1 to 0.9 and find which one gives best F1
thresholds = np.arange(0.1, 0.9, 0.01)
f1_scores  = [f1_score(y_test, y_prob_rf >= t) for t in thresholds]

best_thresh = thresholds[np.argmax(f1_scores)]
print(f"Best Threshold: {best_thresh:.2f} → F1: {max(f1_scores):.4f}")
```

```python
# ============================================================
# SNIPPET 3: COMPARISON — Which model is better?
# Run after Snippet 2A and 2B
# ============================================================

import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, precision_recall_curve

print("=" * 50)
print("MODEL COMPARISON SUMMARY")
print("=" * 50)
print(f"{'Model':<20} {'CV ROC-AUC':>12} {'Test ROC-AUC':>13} {'PR-AUC':>10}")
print("-" * 58)

for name, res in model_results.items():
    print(f"{name:<20} {res['cv_roc_auc']:>12.4f} {res['roc_auc']:>13.4f} {res['pr_auc']:>10.4f}")

# --- Plot ROC Curves Side by Side ---
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

for name, res in model_results.items():
    # ROC Curve
    fpr, tpr, _ = roc_curve(y_test, res["y_prob"])
    axes[0].plot(fpr, tpr, label=f"{name} (AUC={res['roc_auc']:.4f})")

    # Precision-Recall Curve
    prec, rec, _ = precision_recall_curve(y_test, res["y_prob"])
    axes[1].plot(rec, prec, label=f"{name} (PR-AUC={res['pr_auc']:.4f})")

# ROC plot
axes[0].plot([0,1],[0,1],'k--', label="Random Guess")
axes[0].set_title("ROC Curve Comparison")
axes[0].set_xlabel("False Positive Rate")
axes[0].set_ylabel("True Positive Rate")
axes[0].legend()
axes[0].grid(alpha=0.3)

# PR plot
axes[1].set_title("Precision-Recall Curve Comparison")
axes[1].set_xlabel("Recall")
axes[1].set_ylabel("Precision")
axes[1].legend()
axes[1].grid(alpha=0.3)

plt.tight_layout()
plt.show()

# --- Verdict ---
print("\n--- VERDICT ---")
best_roc = max(model_results, key=lambda m: model_results[m]["roc_auc"])
best_pr  = max(model_results, key=lambda m: model_results[m]["pr_auc"])
print(f"Best ROC-AUC  → {best_roc}")
print(f"Best PR-AUC   → {best_pr}  (this matters more for fraud detection!)")
print("\nTIP: For fraud detection, PR-AUC is the more trustworthy metric")
print("because the dataset is heavily imbalanced (~0.17% fraud).")
print("A model that catches more frauds (high recall) without too many")
print("false alarms (high precision) is what you want in production.")
```

```python
# ============================================================
# INTERACTIVE FRAUD CHECKER — Enter a row number, get verdict
# ============================================================
# HOW IT WORKS:
# Your creditcard dataset has 284,807 rows numbered 0 to 284,806.
# You type any row number, and this code pulls that exact transaction,
# runs it through your trained model, and tells you FRAUD or LEGITIMATE.
# It also shows you the actual transaction details so you can see
# what the model is looking at before making its decision.
# ============================================================

import pandas as pd
import numpy as np

# Tell the function which model won your comparison.
# Swap xgb for rf if Random Forest had the better PR-AUC score.
best_model  = xgb
best_thresh = 0.3   # Your tuned threshold from the threshold tuning step

def predict_from_row():

    print("=" * 50)
    print("   FRAUD DETECTION - Row-Based Transaction Check")
    print("=" * 50)
    print(f"   Dataset has {len(X_test)} rows in test set (0 to {len(X_test)-1})")
    print("=" * 50)

    # Ask user to type a row number
    row_num = int(input("\nEnter row number to check: "))

    # Validate the row number is within range
    if row_num < 0 or row_num >= len(X_test):
        print(f"Invalid row number. Please enter between 0 and {len(X_test)-1}")
        return

    # Pull that single row from X_test
    # .iloc[row_num] gets the row by position, just like Excel row numbers
    single_row    = X_test.iloc[[row_num]]   # double brackets keep it as a DataFrame
    actual_label  = y_test.iloc[row_num]     # the real answer: 0 = legit, 1 = fraud

    # Show the transaction details to the user
    print("\n--- Transaction Details ---")
    print(f"  Amount      : ${X_test.iloc[row_num]['Amount']:.2f}")
    print(f"  Hour        : {int(X_test.iloc[row_num]['Hour'])}:00")
    print(f"  V10         : {X_test.iloc[row_num]['V10']:.4f}")
    print(f"  V12         : {X_test.iloc[row_num]['V12']:.4f}")
    print(f"  V14         : {X_test.iloc[row_num]['V14']:.4f}")
    print(f"  V17         : {X_test.iloc[row_num]['V17']:.4f}")
    print(f"  Actual Class: {'FRAUD' if actual_label == 1 else 'LEGITIMATE'} (ground truth)")

    # Run the row through your trained model
    # predict_proba gives two probabilities: [prob_legit, prob_fraud]
    # We take index [0][1] which is the fraud probability
    prob    = best_model.predict_proba(single_row)[0][1]
    verdict = "FRAUD" if prob >= best_thresh else "LEGITIMATE"

    # Display the result
    print("\n" + "=" * 50)
    print(f"  Fraud Probability : {prob*100:.2f}%")
    print(f"  Model Verdict     : *** {verdict} ***")
    print(f"  Actual Answer     : *** {'FRAUD' if actual_label==1 else 'LEGITIMATE'} ***")

    # This line tells you whether the model got it right or wrong
    if verdict == ('FRAUD' if actual_label == 1 else 'LEGITIMATE'):
        print("  Result            : Model was CORRECT")
    else:
        print("  Result            : Model was WRONG")

    print("=" * 50)

# ---- Call the function ----
predict_from_row()
```

```python
!pip install shap -q

import shap

# Initialize the SHAP explainer — TreeExplainer is optimized specifically
# for tree-based models like XGBoost, LightGBM, and Random Forest.
# It's much faster than the generic KernelExplainer for these model types.
explainer = shap.TreeExplainer(xgb)  # swap xgb for rf or lgbm as needed

# Calculate SHAP values for the test set.
# Each value tells us how much each feature contributed to each prediction.
# We take a sample of 500 rows to keep it fast — SHAP on 57k test rows is slow.
sample = X_test.sample(500, random_state=42)
shap_values = explainer.shap_values(sample)

# --- Plot 1: Summary Plot ---
# This shows ALL features ranked by how much they influence predictions overall.
# Red = pushes toward fraud, Blue = pushes away from fraud.
print("SHAP Summary Plot — Which features matter most overall?")
shap.summary_plot(shap_values, sample, plot_type="bar")

# --- Plot 2: Beeswarm Plot ---
# More detailed version — shows the distribution of SHAP values per feature.
# You can see not just which features matter, but HOW they influence predictions.
shap.summary_plot(shap_values, sample)

# --- Plot 3: Single Prediction Explanation ---
# Pick one transaction and explain exactly why the model predicted what it did.
# This is the most intuitive visualization — great for presentations.
print("\nExplaining a single fraud prediction:")
shap.force_plot(
    explainer.expected_value,
    shap_values[0],       # SHAP values for the first sample
    sample.iloc[0],       # The actual feature values for that sample
    matplotlib=True       # Use matplotlib so it renders in Colab
)
```
